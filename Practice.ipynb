```python
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_data = np.arange(1,10,dtype=int)
sample_data = np.concatenate([sample_data, [45,45,67,89,90]])
sample_data
```

```python
# find mean median mode
mean = np.mean(sample_data) 
median = np.median(sample_data)
mode= np.argmax(np.bincount(sample_data))
print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
```

```python
# box plot
import numpy as np
import matplotlib.pyplot as plt

# Sample data


# Create the box plot
plt.boxplot(sample_data, flierprops=dict(marker='o', color='red', markersize=8))

# Add labels and title
plt.xlabel("Sample")
plt.ylabel("Values")
plt.title("Box Plot with Data Points")

# Show the plot
plt.show()

# histogram     
plt.figure(figsize=(10,5))
sns.histplot(sample_data, bins=2, kde=True)
plt.title("Histogram")
plt.xlabel("Sample Data")
plt.ylabel("Frequency") 
plt.show()
```

```python
# imports random module
import random

# Generates a random number between
# a given positive range
r1 = random.randint(0, 10)
print("Random number between 0 and 10 is % s" % (r1))

# Generates a random number between 
# two given negative range
r2 = random.randint(-10, -1)
print("Random number between -10 and -1 is % d" % (r2))

# Generates a random number between 
# a positive and a negative range
r3 = random.randint(-5, 5)
print("Random number between -5 and 5 is % d" % (r3))
```

```python
# Set the style
sns.set_theme(style="whitegrid")

# Create the figure and axis
plt.figure(figsize=(6, 4))

# Boxplot with outliers shown
sns.boxplot(y=sample_data, showfliers=True, color='lightblue')

# Overlay swarmplot to show individual data points
sns.swarmplot(y=sample_data, color='red', size=6)

# Labels and title
plt.xlabel("Sample")
plt.ylabel("Values")
plt.title("Box Plot with Data Points using Seaborn")

# Show plot
plt.show()
```
